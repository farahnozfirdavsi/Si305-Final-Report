```python
# Cleaning the zip codes for each dataframe
broadband_data['zip'] = broadband_data['ZCTA19'].astype(str).str.zfill(5)
healthcare_data['zip'] = healthcare_data['ZCTA20'].astype(str).str.zfill(5)
socio_data['zip'] = socio_data['ZCTA20'].astype(str).str.zfill(5)
```

```python
# filtering for Michigan data in broadband dataset

def filter_michigan(df):
    return df[df["zip"].str.startswith(("48", "49"))]

broadband_data = filter_michigan(broadband_data)
healthcare_data = filter_michigan(healthcare_data)
socio_data = filter_michigan(socio_data)
```

## Key Variables from Each DF

# Merging Dataframes

```python
df = broadband_data.merge(healthcare_data[['zip', 'COUNT_HOSPITALS']], on= 'zip', how = 'inner')
df = df.merge(socio_data[['zip', 'DISADVANTAGE16_20']], on = 'zip', how='inner')

df['DISADVANTAGE16_20'] = pd.to_numeric(df['DISADVANTAGE16_20'], errors='coerce')
df['ses_quartile'] = pd.qcut(df['DISADVANTAGE16_20'], 4, labels = ['Low', 'Medium', 'High', 'Very High'])
```
